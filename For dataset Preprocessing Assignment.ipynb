```python
quan=[]
qual=[]
for columnName in dataset.columns:
   # print(columnName)
    if(dataset[columnName].dtype=='O'):
        #print("qual")
        qual.append(columnName)
    else:
       # print("quan")
        quan.append(columnName)
```

```python
quan
```

```python
qual
```

```python
dataset[quan]
```

```python
dataset[qual]
```

```python
#missing values replace dataset same time multiple column
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```

```python
df
```

```python
#create table 
df = pd.DataFrame(df,columns = quan)
```

```python
df
```

```python
#missing values replace dataset same time multiple column
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imp.fit(dataset[qual])
dff=imp.transform(dataset[qual])
```

```python
#create table 
dff = pd.DataFrame(dff,columns = qual)
```
